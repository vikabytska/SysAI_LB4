```python
import numpy as np
from sklearn import linear_model
import sklearn.metrics as sm
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
```

```python
np.random.seed(0)
m = 100
```

```python
X = 6 * np.random.rand(m, 1) - 5
y = 0.5 * X ** 2 + X + 2 + np.random.randn(m, 1)
```

```python
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)
```

```python
# Лінійна регресія
linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(X, y)
y_linear_pred = linear_regressor.predict(X)

# Поліноміальна регресія
polynomial = PolynomialFeatures(degree=2, include_bias=False)
X_poly = polynomial.fit_transform(X)
poly_linear_model = linear_model.LinearRegression()
poly_linear_model.fit(X_poly, y)
y_poly_pred = poly_linear_model.predict(X_poly)

print("\nR2 score for polynomial regression:", sm.r2_score(y, y_poly_pred))
```

```python
poly_linear_model = linear_model.LinearRegression()
poly_linear_model.fit(X_poly, y)
y_pred = poly_linear_model.predict(X_poly)

print("\nr2: ", sm.r2_score(y, y_pred))
```

```python
# Лінійна регресія
plt.scatter(X, y, color='blue')
plt.plot(X, linear_regressor.predict(X), color='red', linewidth=1)
plt.title("Лінійна регресія")
plt.show()
```

```python
# Поліноміальна регресія
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, "*", color='red', linewidth=2)
plt.title("Поліноміальна регресія")
plt.show()
```

```python
# Метрики для лінійної регресії
print("Linear regressor performance:")
print("Mean absolute error =", round(sm.mean_absolute_error(y, y_linear_pred), 2))
print("Mean squared error =", round(sm.mean_squared_error(y, y_linear_pred), 2))
print("Median absolute error =", round(sm.median_absolute_error(y, y_linear_pred), 2))
print("Explained variance score =", round(sm.explained_variance_score(y, y_linear_pred), 2))
print("R2 score =", round(sm.r2_score(y, y_linear_pred), 2))
```

```python
# Метрики для поліноміальної регресії
print("\nPolynomial regressor performance:")
print("Mean absolute error =", round(sm.mean_absolute_error(y, y_poly_pred), 2))
print("Mean squared error =", round(sm.mean_squared_error(y, y_poly_pred), 2))
print("Median absolute error =", round(sm.median_absolute_error(y, y_poly_pred), 2))
print("Explained variance score =", round(sm.explained_variance_score(y, y_poly_pred), 2))
print("R2 score =", round(sm.r2_score(y, y_poly_pred), 2))
```

```python
# Коефіцієнти поліноміальної регресії
print('\nPolynomial regression coefficients:')
print('Coefficients:', poly_linear_model.coef_)
print('Intercept:', poly_linear_model.intercept_)
```
