```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([6, 5, 7, 10])
```

```python
# Створюємо матрицю X: перший стовпчик — це одиниці (для β0), другий — значення x (для β1)
X = np.column_stack((np.ones(len(x)), x))

# Розв'язуємо систему нормальних рівнянь, щоб знайти β0 і β1
beta = np.linalg.inv(X.T @ X) @ X.T @ y
```

```python
# Отримані коефіцієнти
beta_0, beta_1 = beta
print(f"Рівняння прямої: y = {beta_0:.2f} + {beta_1:.2f}x")
```

```python
# Обчислення мінімальної суми квадратів похибок
y_pred = beta_0 + beta_1 * x
squared_errors = (y - y_pred)**2
S = np.sum(squared_errors)
print(f"Мінімальна сума квадратів похибок: {S:.2f}")
```

```python
# Візуалізація
plt.scatter(x, y, color='red', label='Точки')
plt.plot(x, y_pred, label=f'y = {beta_0:.2f} + {beta_1:.2f}x', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Метод найменших квадратів')
plt.legend()
plt.grid(True)
plt.show()
```
